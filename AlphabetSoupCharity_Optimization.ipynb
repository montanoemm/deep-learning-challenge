```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf


```

```python
#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

## # First Optimization
*We will keep the column NAME to test its relevance

```python
# Drop the non-beneficial ID columns, 'EIN' 
#  YOUR CODE GOES HERE
application_df = application_df.drop(columns = ['EIN'])
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
application_df.nunique()
```

```python
#Look at NAME value counts for binning
name_vc= application_df["NAME"].value_counts()
name_vc
```

```python
#Choose a cutoff value and create a list of name_replace to be replaced
#use variable name: name_replace
name_replace= list(name_vc[name_vc<10].index)
name_replace
```

```python
#Raplace names in the dataframe for any name repeated less than 10 times
for name in name_replace:
    application_df["NAME"]= application_df["NAME"].replace(name,"Other")

#Check name binning was succesful
application_df["NAME"].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
app_type_binning=application_df["APPLICATION_TYPE"].value_counts()
app_type_binning
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Chosen value 500
#  YOUR CODE GOES HERE
application_types_to_replace= list(app_type_binning[app_type_binning<500].index)
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
classification_binning= application_df['CLASSIFICATION'].value_counts()
classification_binning
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
classification_binning_biggerthan1=classification_binning[classification_binning>1]
classification_binning_biggerthan1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Chosen random value of 100
#  YOUR CODE GOES HERE
classifications_to_replace= list(classification_binning[classification_binning<100].index)


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
print(application_df.dtypes)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
application_df_dummies= pd.get_dummies(application_df, columns=['NAME', 'APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'])
application_df_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
# Features for training "X"
X = application_df_dummies.drop("IS_SUCCESSFUL", axis=1).values

# Target for the training "y"

y = application_df_dummies["IS_SUCCESSFUL"].values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train.shape
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
number_input_features= len(X_train_scaled[0])
hidden_layer_node1=2
hidden_layer_node2=4

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_layer_node1, activation="relu", input_dim=number_input_features))
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_layer_node2, activation="relu"))

# Output layer only one output
#  YOUR CODE GOES HERE
# Sigmoid is used for the output because is a commonly used activation function for binary classification tasks,
#  as it produces output values in the range [0, 1], which can be interpreted as probabilities.
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Second Optimization: adding more nodes

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  After improving the accuracy of the previous model through the manipulation of data
# more nodes were added to to test the newa accuracy of the model.
number_input_features= len(X_train_scaled[0])
hidden_layer_node1=8
hidden_layer_node2=16

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_layer_node1, activation="relu", input_dim=number_input_features))
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_layer_node2, activation="relu"))

# Output layer only one output
#  YOUR CODE GOES HERE
# Sigmoid is used for the output because is a commonly used activation function for binary classification tasks,
#  as it produces output values in the range [0, 1], which can be interpreted as probabilities.
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
